```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
flights = sns.load_dataset("flights")
```

```python
corr_matrix = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

```python
sns.pairplot(iris, hue='species', palette='husl',
              diag_kind='kde', markers=['o', 's', 'D'], plot_kws={'alpha': 0.7})

plt.suptitle("     Iris", y=1.02)
plt.show()
```

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(12, 8))
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)

plt.title('      ')
plt.xlabel('')
plt.ylabel('')

plt.show()
```
