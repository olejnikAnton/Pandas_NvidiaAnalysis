```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
penguins.head
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, kde=True, color='blue')

plt.title('Распределение длины клюва у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Частота')

plt.show()
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, kde=True, hue='species')

plt.title('Распределение длины клюва у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Частота')

plt.show()
```

```python
species_counts = penguins['species'].value_counts()

plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Распределение пингвинов по видам')

# Убедимся, что круг круглый
plt.axis('equal')

plt.show()
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species')

plt.title('Длина клюва и плавников у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавников (мм)')
plt.legend(title='Вид пингвина')


plt.show()
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm')

plt.title('Длина клюва и плавников у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')

plt.show()
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', 
           hue='species', legend_out=False)


plt.title('Длина клюва и плавников у пингвинов')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')
plt.legend(title='Виды пингвинов')

plt.show()
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', order=2, ci=None)

plt.title('Взаимосвязь длины клюва и плавников')
plt.xlabel('Длина клюва (мм)')
plt.ylabel('Длина плавника (мм)')

plt.show()
```
